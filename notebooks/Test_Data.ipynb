```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

data_slice = dataframe.iloc[0:800]
fig = make_subplots(rows=1, cols=len(data_slice.columns[5:11]), shared_yaxes=False)

for i, column in enumerate(data_slice.columns[5:11], start=1):
    fig.add_trace(
        go.Scatter(x=data_slice["No"], y=data_slice[column], mode="lines"), row=1, col=i
    )
    fig.update_yaxes(title_text=column, row=1, col=i)

fig.update_layout(showlegend=False)
fig.show()
```
